## Eliminar guión atributo "rut"

```python
Rdig = df2['rut'].astype(str).str[-1:] #devuelve el ultimo digito
Rdig
```

```python
Rrut = df2['rut'].astype(str).str[:-2]
Rrut
```

```python
R = Rrut+""+Rdig
R
```

```python
df2['rut'] = R.astype(int) # asignamos al atributo rut la variable R, tipo entero
```

```python
df = pd.merge(df1,df2, left_on='rut', right_on='rut') #funcion merge: unir
df
```

```python
df = pd.merge(df,df3, left_on='rut', right_on='rut')
df
```

Eliminar atributos repetidos

```python
df = df.drop(labels=["nombre_completo","primer_nombre_y","primer_apellido_y"], axis=1)
df
```

Atributo "sexo_y"

```python
df['sexo_y'] = df['sexo_y'].replace(1,'F')
df['sexo_y'] = df['sexo_y'].replace(0,'M')
df
```

Unir los atributos "sexo_x" y "sexo_y"

```python
ynan = np.where(df['sexo_x']==np.nan, df['sexo_y'], df['sexo_x'])
df['sexo'] = np.where(df['sexo_x']==df['sexo_y'], df['sexo_x'], ynan)
df
```

```python
df = df.drop(labels=["sexo_x","sexo_y"], axis=1)
df
```

```python
df['sexo'] = df['sexo'].fillna(' ') #eliminar datos NA del atributo "sexo"
df
```

# Atributo "edad"

```python
df.dtypes
```

```python
df['edad_x'] = df['edad_x'].astype(str).str[:-2]
df['edad_y'] = df['edad_y'].astype(str).str[:-2]
df
```

```python
y_nan = np.where(df['edad_x']=='n', df['edad_y'], df['edad_x'])
df['edad_x'] = np.where(df['edad_x']==df['edad_y'], df['edad_x'], y_nan)
df
```

```python
y_nan_ = np.where(df['edad_x']=='n', df['edad'], df['edad_x']+' años')
df['edad'] = np.where(df['edad_x']+' años'==df['edad'], df['edad'], y_nan_)
df
```

```python
df = df.drop(labels=["edad_x","edad_y"], axis=1)
df
```

```python
df.columns = ['Id','Nombre_1','Nombre_2','Apellido_1','Apellido_2','Edad','Genero']
df
#renombrar atributos
```
