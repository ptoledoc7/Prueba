```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ancho = 12 pulgadas
# Largo = 6 pulgadas
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df_titanic = sns.load_dataset("titanic")
```

```python
df_titanic.shape
```

```python
df_titanic.head(5)
```

```python
total_clases = df_titanic.groupby("class").size()
print(total_clases)
```

```python
nom_clases = "Firts", "Second", "Third"
grafico_clases = plt.pie(total_clases, labels = nom_clases, autopct='%1.1f%%')
plt.title("Pasajeros del Titanic por clase")

plt.show()
```

```python
total_survived = df_titanic.groupby(["survived", "class"]).size()
print(total_survived)
```

```python
fig, charts = plt.subplots(1,2) # Orden del grafico (filas, columnas)

# Definiendo cada grfico

charts[0].pie(total_survived[0], labels = nom_clases, autopct='%1.1f%%')
charts[1].pie(total_survived[1], labels = nom_clases, autopct='%1.1f%%')

# Titulos de cada grafico
charts[0].set_title("No sobrevivientes")
charts[1].set_title("Sobrevivientes")

# Nombre general de los graficos
fig.suptitle("Pasajeros del Titanic por Supervivencia (0/1) y por Clase (T/F/S)")

plt.show()
```
