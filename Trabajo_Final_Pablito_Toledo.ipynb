Revisando las tablas, se puede observar que existe una columna llave (ID), la cual es una dato entero en todas las tablas, para lo cual se puede proceder a unir las tablas

```python
# Uniendo tablas
df = pd.merge(df1,df2, left_on='ID', right_on='ID')
df = pd.merge(df,df3, left_on='ID', right_on='ID')
df4 = df4.drop('Unnamed: 0', axis=1)
df = pd.merge(df,df4, left_on='ID', right_on='ID')
df
```

```python
df.dtypes
```

```python
df = df.drop(labels=["Unnamed: 0_x"], axis=1)
df = df.drop(labels=["Unnamed: 0_y"], axis=1)
df
```

```python
df.dtypes
```

```python
ynan = np.where(df['LIMIT_BAL_x']==np.nan, df['LIMIT_BAL_y'], df['LIMIT_BAL_x'])
df['LIMIT_BAL'] = np.where(df['LIMIT_BAL_x']==df['LIMIT_BAL_y'], df['LIMIT_BAL_x'], ynan)
df
```

```python
df.drop(['LIMIT_BAL_x', 'LIMIT_BAL_y'], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
# Analizando Edad (AGE)
df['AGE_y'] = df['AGE_y'].astype(str).str[:-2]
df
```

```python
y_nan_ = np.where(df['AGE_y']=='n', df['AGE_x'], df['AGE_y']+' años')
df['AGE_x'] = np.where(df['AGE_y']+' años'==df['AGE_x'], df['AGE_x'], y_nan_)
df
```

```python
df = df.drop(labels=["AGE_y"], axis=1)
df
```

```python
# Analizando Educacion (EDUCATION)
df['EDUCATION_y'] = df['EDUCATION_y'].astype(str).str[:-2]
df
```

```python
z_nan_ = np.where(df['EDUCATION_y']=='n', df['EDUCATION_x'], df['EDUCATION_y'])
df['EDUCATION_x'] = np.where(df['EDUCATION_y']==df['EDUCATION_x'], df['EDUCATION_x'], z_nan_)
df
```

```python
df = df.drop(labels=["EDUCATION_y"], axis=1)
df
```

```python
# Analizando SEXO
df['SEX_y'] = df['SEX_y'].replace(1,'male')
df['SEX_y'] = df['SEX_y'].replace(2,'female')
df
```

```python
x_nan_ = np.where(df['SEX_y']=='n', df['SEX_x'], df['SEX_y'])
df['SEX_x'] = np.where(df['SEX_y']==df['SEX_x'], df['SEX_x'], x_nan_)
df
```

```python
df = df.drop(labels=["SEX_y"], axis=1)
df
```

```python
# Aplicando nombre a los generos de SEXO
df['SEX_x'] = df['SEX_x'].replace('1','male')
df['SEX_x'] = df['SEX_x'].replace('2','female')
df
```

```python
df = df.drop(labels=["Unnamed: 0"], axis=1)
df
```

```python
# Elimando registros duplicados
df.drop_duplicates(subset=None, keep='first', inplace=True)
df
```

```python
df.info()
```

```python
df = df.rename(columns={'SEX_x': 'SEX', 'AGE_x': 'AGE', 'EDUCATION_x': 'EDUCATION'})
df.head(5)
```
